```python
from chembl_webresource_client.new_client import new_client

def get_ic50_from_chembl_by_inchikey(inchikey):

    molecule = new_client.molecule
    res = molecule.filter(inchi_key=inchikey).only(['molecule_chembl_id'])
    
    if not res:
        return None, None
    
    chembl_id = res[0]['molecule_chembl_id']

    activity = new_client.activity
    acts = activity.filter(
        molecule_chembl_id=chembl_id,
        target_chembl_id="CHEMBL4630844",
        standard_type__in=["IC50", "Ki"],
        standard_units="nM",
        standard_value__isnull=False
    ).only(['standard_value', 'standard_type', 'standard_units'])
    
    if acts:
        val = float(acts[0]['standard_value'])
        typ = acts[0]['standard_type']
        return val, typ
    return None, None
```

```python
results = []
for ligand, inchikey in ligand_inchikey.items():
    ic50, typ = get_ic50_from_chembl_by_inchikey(inchikey)
    results.append({
        'Ligand': ligand,
        'InChIKey': inchikey,
        'IC50_nM': ic50,
        'Type': typ
    })
    print(f"{ligand}: {ic50} nM")

chembl_df = pd.DataFrame(results)
```

```python
import requests

def get_molecule_by_inchikey(inchikey):

    url = f"https://www.ebi.ac.uk/chembl/api/data/molecule/{inchikey}.json"
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        return {
            'molecule_chembl_id': data.get('molecule_chembl_id'),
            'pref_name': data.get('pref_name')
        }
    return None

def get_activities_by_chembl_id(chembl_id):

    url = f"https://www.ebi.ac.uk/chembl/api/data/activity.json?molecule_chembl_id={chembl_id}&limit=100"
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        return data.get('activities', [])
    return []


inchikey = "WFDCRJQKOODGGI-GFCCVEGCSA-N"

# Шаг 1: получаем молекулу
mol = get_molecule_by_inchikey(inchikey)
if not mol:
    print("❌ Молекула не найдена")
else:
    print("✅ Молекула найдена:", mol)
    chembl_id = mol['molecule_chembl_id']
    
    # Шаг 2: получаем активности
    activities = get_activities_by_chembl_id(chembl_id)
    
    if not activities:
        print("❌ Нет активностей в ChEMBL")
    else:
        print(f"✅ Найдено {len(activities)} активностей. Примеры:")
        for act in activities[:5]:  # первые 5
            print({
                'target': act.get('target_chembl_id'),
                'target_name': act.get('target_pref_name'),
                'type': act.get('standard_type'),
                'value': act.get('standard_value'),
                'units': act.get('standard_units')
            })
```

```python
import requests
import pandas as pd

def get_ic50_from_chembl_by_inchikey(inchikey):

    mol_url = f"https://www.ebi.ac.uk/chembl/api/data/molecule/{inchikey}.json"
    mol_resp = requests.get(mol_url)
    
    if mol_resp.status_code != 200:
        return None, None, None
    
    mol_data = mol_resp.json()
    chembl_id = mol_data.get('molecule_chembl_id')
    if not chembl_id:
        return None, None, None

    irak4_ids = ["CHEMBL3351", "CHEMBL4630844", "CHEMBL3778", "CHEMBL3357"]
    
    for target_id in irak4_ids:
        act_url = (
            f"https://www.ebi.ac.uk/chembl/api/data/activity.json"
            f"?molecule_chembl_id={chembl_id}"
            f"&target_chembl_id={target_id}"
            f"&standard_type__in=IC50,Ki,Kd,EC50,Potency"
            f"&standard_value__isnull=false"
            f"&limit=10"
        )
        act_resp = requests.get(act_url)
        if act_resp.status_code == 200:
            act_data = act_resp.json()
            activities = act_data.get('activities', [])
            if activities:
                act = activities[0]
                value_str = act.get('standard_value')
                unit = act.get('standard_units')
                typ = act.get('standard_type')
                
                if not value_str or not unit:
                    continue
                
                try:
                    value = float(value_str)
                except ValueError:
                    continue

                # Конвертация в нМ
                if unit == 'nM':
                    ic50_nm = value
                elif unit in ('uM', 'µM'):
                    ic50_nm = value * 1000
                elif unit == 'mM':
                    ic50_nm = value * 1_000_000
                elif unit == 'pM':
                    ic50_nm = value / 1000
                else:
                    continue

                return ic50_nm, typ, unit

    return None, None, None
```

```python
results = []
for ligand, inchikey in ligand_inchikey.items():
    ic50_nm, typ, unit = get_ic50_from_chembl_by_inchikey(inchikey)
    results.append({
        'Ligand': ligand,
        'InChIKey': inchikey,
        'IC50_nM': ic50_nm,
        'Type': typ,
        'Original_Unit': unit
    })
    if ic50_nm is not None:
        print(f"{ligand}: {ic50_nm:.2f} nM")
    else:
        print(f"{ligand}: не найдено")

chembl_df = pd.DataFrame(results)
```
