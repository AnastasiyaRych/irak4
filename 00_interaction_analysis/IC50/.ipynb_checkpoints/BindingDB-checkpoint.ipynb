```python
bd = pd.read_csv("bindingdb_irak4.tsv", sep="\t", low_memory=False)

if 'Ligand InChI Key' not in bd.columns:
    print("❌ В файле BindingDB нет колонки 'Ligand InChI Key'")
else:

    result = pd.merge(
        inchikey_df,
        bd[['Ligand InChI Key', 'IC50 (nM)', 'Ki (nM)', 'Kd (nM)']],
        left_on='InChIKey',
        right_on='Ligand InChI Key',
        how='inner'
    )

    print(f"\n✅ Найдено совпадений: {len(result)}")
    
    if len(result) > 0:
        # Выбираем лучшую доступную аффинность (IC50 > Ki > Kd)
        def get_affinity(row):
            if pd.notna(row['IC50 (nM)']):
                return row['IC50 (nM)'], 'IC50'
            elif pd.notna(row['Ki (nM)']):
                return row['Ki (nM)'], 'Ki'
            elif pd.notna(row['Kd (nM)']):
                return row['Kd (nM)'], 'Kd'
            return None, None

        result[['Affinity_nM', 'Affinity_type']] = result.apply(
            get_affinity, axis=1, result_type='expand'
        )
        
        # Покажем результат
        display_cols = ['Ligand', 'InChIKey', 'Affinity_nM', 'Affinity_type']
        print(result[display_cols])
    else:
        print("❌ Совпадений не найдено. Проверьте, что InChIKey введены верно.")
```
